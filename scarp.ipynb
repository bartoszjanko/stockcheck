```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.gpw.pl/spolki"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Znajdź tabelę (może wymagać dopasowania selektora)
table = soup.find('table', {'class': 'table'})  # Sprawdź, jaka jest rzeczywista klasa tabeli

# Jeśli tabela jest ładowana dynamicznie (np. przez JavaScript), requests nie wystarczy - użyj selenium
if not table:
    print("Tabela nie została znaleziona. Strona może wymagać Selenium.")

# Przetwarzanie wierszy
data = []
for row in table.find_all('tr')[1:]:  # Pomijamy nagłówek
    cols = row.find_all('td')
    if len(cols) >= 3:
        name = cols[0].get_text(strip=True)
        market = cols[1].get_text(strip=True)
        sectors = cols[2].get_text(strip=True)
        data.append([name, market, sectors])

# Tworzenie DataFrame (opcjonalnie)
df = pd.DataFrame(data, columns=['Nazwa spółki', 'Rynek', 'Sektory'])
print(df.head())
```
